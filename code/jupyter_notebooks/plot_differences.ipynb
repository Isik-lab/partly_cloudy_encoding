<h1>Plot inter-age group differences</h1>

<h2>Setup</h2>

```python
# Functional imports
import os 
import numpy as np
import nibabel as nib
from nilearn.image import threshold_img

from data_prep_helpers import make_vol, calculate_difference
from plotting_helpers import plot_lateral_only, plot_lateral_only_with_vmax, plot_surfaces_with_vmax

# Notebook specific imports
import warnings
import sys 
if not sys.warnoptions:
    warnings.simplefilter("ignore")
import matplotlib.pyplot as plt
%autosave 5
```

```python
# Set data location
directory = os.path.join(os.path.expanduser('~'), 'Documents', 'Data')
print('Data location:', directory)

# Load mask
mask = nib.load(os.path.join(directory, 'group_gm_mask.nii.gz'))
print('Mask:', np.shape(mask))
```

<h2>Adult thresholded encoding results</h2>

```python
path = os.path.join(directory, 'adults', 'encoding_results', 'thresholded')
full_thresholded_adults = np.load(os.path.join(path, 'adults_full_scores_gm_mask_thresholded05.npy'))
moten_thresholded_adults = np.load(os.path.join(path, 'adults_moten_only_scores_gm_mask_thresholded05.npy'))
social_thresholded_adults = np.load(os.path.join(path, 'adults_social_only_scores_gm_mask_thresholded05.npy'))
```

```python
plot_lateral_only_with_vmax(vol=make_vol(full_thresholded_adults, mask), 
                            title='Full scores - adults', 
                            vis_threshold=0.01, vmax=0.5)

plot_lateral_only_with_vmax(vol=make_vol(moten_thresholded_adults, mask), 
                            title='Motion energy scores - adults', 
                            vis_threshold=0.01, vmax=0.5)

plot_lateral_only_with_vmax(vol=make_vol(social_thresholded_adults, mask), 
                            title='Social feature scores - adults', 
                            vis_threshold=0.01, vmax=0.3)
```

<h2>Comparisons</h2>

```python
# Load adults
path = os.path.join(directory, 'adults', 'encoding_results', 'unthresholded')
full_adults = np.load(os.path.join(path, 'adults_full_scores_gm_mask.npy'))
moten_adults = np.load(os.path.join(path, 'adults_moten_only_scores_gm_mask.npy'))
social_adults = np.load(os.path.join(path, 'adults_social_only_scores_gm_mask.npy'))
```

<h3>3-4 year olds</h3>

```python
# 3-4 years
path = os.path.join(directory, '3-4-yrs', 'encoding_results', 'unthresholded')
full_diff_34yrs = calculate_difference(group1=full_adults, 
                                       group2=np.load(os.path.join(path, '3-4-yrs_full_scores_gm_mask.npy')))
moten_diff_34yrs = calculate_difference(group1=moten_adults, 
                                        group2=np.load(os.path.join(path, '3-4-yrs_moten_only_scores_gm_mask.npy')))
social_diff_34yrs = calculate_difference(group1=social_adults, 
                                         group2=np.load(os.path.join(path, '3-4-yrs_social_only_scores_gm_mask.npy')))

plot_lateral_only_with_vmax(vol=make_vol(full_diff_34yrs, mask), 
                            title='Full scores difference - 3-4 years', 
                            vis_threshold=0.1, vmax=0.25)

plot_lateral_only_with_vmax(vol=make_vol(moten_diff_34yrs, mask), 
                            title='Motion energy scores difference - 3-4 years', 
                            vis_threshold=0.1, vmax=0.25)

plot_lateral_only_with_vmax(vol=make_vol(social_diff_34yrs, mask), 
                            title='Social feature scores difference - 3-4 years', 
                            vis_threshold=0.1, vmax=0.25)
```

```python
clustered_full_diff_34yrs = threshold_img(img=make_vol(full_diff_34yrs, mask),
                                          threshold=0.0, cluster_threshold=10, two_sided=False)

plot_lateral_only_with_vmax(vol=clustered_full_diff_34yrs, 
                            title='Clustered full scores difference - 3-4 years', 
                            vis_threshold=0.01, vmax=0.2)

clustered_moten_diff_34yrs = threshold_img(img=make_vol(moten_diff_34yrs, mask),
                                          threshold=0.0, cluster_threshold=10, two_sided=False)

plot_lateral_only_with_vmax(vol=clustered_moten_diff_34yrs, 
                            title='Clustered motion energy scores difference - 3-4 years', 
                            vis_threshold=0.01, vmax=0.2)

clustered_social_diff_34yrs = threshold_img(img=make_vol(social_diff_34yrs, mask),
                                          threshold=0.0, cluster_threshold=10, two_sided=False)

plot_lateral_only_with_vmax(vol=clustered_social_diff_34yrs, 
                            title='Clustered social feature scores difference - 3-4 years', 
                            vis_threshold=0.01, vmax=0.2)
```
