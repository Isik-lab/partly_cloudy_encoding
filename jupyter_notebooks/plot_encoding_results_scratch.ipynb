```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import ttest_rel
```

```python
dataframe = pd.DataFrame()

for subject in within_group_34yrs_unthresholded:
    row = {'Prediction performance': [subject], 'Age group': ['3-4 yrs'], 'Mask used': ['within group']}
    dataframe = pd.concat([dataframe, pd.DataFrame(row)], ignore_index=True, axis=0)

for subject in adult_mask_34yrs_unthresholded:
    row = {'Prediction performance': [subject], 'Age group': ['3-4 yrs'], 'Mask used': ['adult mask']}
    dataframe = pd.concat([dataframe, pd.DataFrame(row)], ignore_index=True, axis=0)

for subject in within_group_5yrs_unthresholded:
    row = {'Prediction performance': [subject], 'Age group': ['5 yrs'], 'Mask used': ['within group']}
    dataframe = pd.concat([dataframe, pd.DataFrame(row)], ignore_index=True, axis=0)

for subject in adult_mask_5yrs_unthresholded:
    row = {'Prediction performance': [subject], 'Age group': ['5 yrs'], 'Mask used': ['adult mask']}
    dataframe = pd.concat([dataframe, pd.DataFrame(row)], ignore_index=True, axis=0)

for subject in within_group_7yrs_unthresholded:
    row = {'Prediction performance': [subject], 'Age group': ['7 yrs'], 'Mask used': ['within group']}
    dataframe = pd.concat([dataframe, pd.DataFrame(row)], ignore_index=True, axis=0)

for subject in adult_mask_7yrs_unthresholded:
    row = {'Prediction performance': [subject], 'Age group': ['7 yrs'], 'Mask used': ['adult mask']}
    dataframe = pd.concat([dataframe, pd.DataFrame(row)], ignore_index=True, axis=0)

for subject in within_group_812yrs_unthresholded:
    row = {'Prediction performance': [subject], 'Age group': ['8-12 yrs'], 'Mask used': ['within group']}
    dataframe = pd.concat([dataframe, pd.DataFrame(row)], ignore_index=True, axis=0)

for subject in adult_mask_812yrs_unthresholded:
    row = {'Prediction performance': [subject], 'Age group': ['8-12 yrs'], 'Mask used': ['adult mask']}
    dataframe = pd.concat([dataframe, pd.DataFrame(row)], ignore_index=True, axis=0)
```

```python
t_34yrs, p_34yrs = ttest_rel(a=adult_mask_34yrs_unthresholded,
                             b=within_group_34yrs_unthresholded, nan_policy='omit', alternative='less')
print(f'3-4 years t: {t_34yrs:.3f}, p: {p_34yrs:.6f}')

t_5yrs, p_5yrs = ttest_rel(a=adult_mask_5yrs_unthresholded,
                           b=within_group_5yrs_unthresholded, nan_policy='omit', alternative='less')
print(f'5 years t: {t_5yrs:.3f}, p: {p_5yrs:.6f}')

t_7yrs, p_7yrs = ttest_rel(a=adult_mask_7yrs_unthresholded,
                           b=within_group_7yrs_unthresholded, nan_policy='omit', alternative='less')
print(f'7 years t: {t_7yrs:.3f} p: {p_7yrs:.6f}')

t_812yrs, p_812yrs = ttest_rel(a=adult_mask_812yrs_unthresholded,
                               b=within_group_812yrs_unthresholded, nan_policy='omit', alternative='less')
print(f'8-12 years t: {t_812yrs:.3f}, p: {p_812yrs:.6f}')

fig = plt.figure(figsize=(14,5))
plt.title('Average Encoding Performance')
ax = sns.boxplot(x="Age group", y="Prediction performance", hue="Mask used", data=dataframe, palette="Set3")
plt.legend(loc='upper left')
plt.show()
```
