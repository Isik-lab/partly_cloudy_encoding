<h3>Average prediction performance</h3>

```python
t_full_34yrs, p_full_34yrs = ttest_ind(a=full_34yrs_unthresholded,b=full_adults_unthresholded)
t_moten_34yrs, p_moten_34yrs = ttest_ind(a=moten_34yrs_unthresholded,b=moten_adults_unthresholded)
t_social_34yrs, p_social_34yrs = ttest_ind(a=social_34yrs_unthresholded,b=social_adults_unthresholded)

t_comparison_34yrs, p_comparison_34yrs = ttest_rel(a=full_34yrs_unthresholded,b=social_34yrs_unthresholded)

print('3-4 years compared to adults (unpaired):')
print(f'Full t: {t_full_34yrs:.3f}, p: {p_full_34yrs:.3f}')
print(f'Moten t: {t_moten_34yrs:.3f}, p: {p_moten_34yrs:.3f}')
print(f'Social t: {t_social_34yrs:.3f}, p: {p_social_34yrs:.3f}')
print('')
print('3-4 years between model comparison (paired):')
print(f'Full vs social t: {t_comparison_34yrs:.3f}, p: {p_comparison_34yrs:.9f}')
print('-')

t_full_5yrs, p_full_5yrs = ttest_ind(a=full_5yrs_unthresholded,b=full_adults_unthresholded)
t_moten_5yrs, p_moten_5yrs = ttest_ind(a=moten_5yrs_unthresholded,b=moten_adults_unthresholded)
t_social_5yrs, p_social_5yrs = ttest_ind(a=social_5yrs_unthresholded,b=social_adults_unthresholded)

t_comparison_5yrs, p_comparison_5yrs = ttest_rel(a=full_5yrs_unthresholded,b=social_5yrs_unthresholded)

print('5 years compared to adults (unpaired):')
print(f'Full t: {t_full_5yrs:.3f}, p: {p_full_5yrs:.3f}')
print(f'Moten t: {t_moten_5yrs:.3f}, p: {p_moten_5yrs:.3f}')
print(f'Social t: {t_social_5yrs:.3f}, p: {p_social_5yrs:.3f}')
print('')
print('5 years between model comparison (paired):')
print(f'Full vs social t: {t_comparison_5yrs:.3f}, p: {p_comparison_5yrs:.9f}')
print('-')

t_full_7yrs, p_full_7yrs = ttest_ind(a=full_7yrs_unthresholded,b=full_adults_unthresholded)
t_moten_7yrs, p_moten_7yrs = ttest_ind(a=moten_7yrs_unthresholded,b=moten_adults_unthresholded)
t_social_7yrs, p_social_7yrs = ttest_ind(a=social_7yrs_unthresholded,b=social_adults_unthresholded)

t_comparison_7yrs, p_comparison_7yrs = ttest_rel(a=full_7yrs_unthresholded,b=social_7yrs_unthresholded)

print('7 years compared to adults (unpaired):')
print(f'Full t: {t_full_7yrs:.3f}, p: {p_full_7yrs:.3f}')
print(f'Moten t: {t_moten_7yrs:.3f}, p: {p_moten_7yrs:.3f}')
print(f'Social t: {t_social_7yrs:.3f}, p: {p_social_7yrs:.3f}')
print('')
print('7 years between model comparison (paired):')
print(f'Full vs social t: {t_comparison_7yrs:.3f}, p: {p_comparison_7yrs:.9f}')
print('-')

t_full_812yrs, p_full_812yrs = ttest_ind(a=full_812yrs_unthresholded,b=full_adults_unthresholded)
t_moten_812yrs, p_moten_812yrs = ttest_ind(a=moten_812yrs_unthresholded,b=moten_adults_unthresholded)
t_social_812yrs, p_social_812yrs = ttest_ind(a=social_812yrs_unthresholded,b=social_adults_unthresholded)

t_comparison_812yrs, p_comparison_812yrs = ttest_rel(a=full_812yrs_unthresholded,b=social_812yrs_unthresholded)

print('8-12 years compared to adults (unpaired):')
print(f'Full t: {t_full_812yrs:.3f}, p: {p_full_812yrs:.3f}')
print(f'Moten t: {t_moten_812yrs:.3f}, p: {p_moten_812yrs:.3f}')
print(f'Social t: {t_social_812yrs:.3f}, p: {p_social_812yrs:.3f}')
print('')
print('8-12 years between model comparison (paired):')
print(f'Full vs social t: {t_comparison_812yrs:.3f}, p: {p_comparison_812yrs:.9f}')
print('-')

t_comparison_adults, p_comparison_adults = ttest_rel(a=full_adults_unthresholded,b=social_adults_unthresholded)

print('Adults between model comparison (paired):')
print(f'Full vs social t: {t_comparison_adults:.3f}, p: {p_comparison_adults:.9f}')
print('-')
```
